```python
import numpy as np
import pandas as pd
np.random.seed(50)

def random_walk(n_steps):
    step= 0.5 - np.random.random(n_steps)
    return np.cumsum(step)

date =pd.date_range('1/1/2019','12/31/2019',freq='D')
data = {'a':random_walk(365), 'b':random_walk(365)}
df=pd.DataFrame(data,index=date)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
flights = pd.read_csv("seaborn/flights.csv")
```

```python
#barplot using titanic data comparing survived and class fields
titanic=pd.read_csv("seaborn/titanic.csv")
sns.barplot(x="class",y="survived",data=titanic)
plt.show()
```

```python
# box plot data comparing age and class fields
sns.boxplot(x="class",y="age",data=titanic)
plt.show()
```

```python
# violin plot data comparing age and class fields
sns.violinplot(x="class",y="age",data=titanic)
plt.show()
```

```python
# heat map comparing month and Year fields
flights = flights.pivot("month","year","passengers")
sns.heatmap(flights)
sns.heatmap(flights,cmap="YlGnBu")
plt.show()
```

```python
import numpy as np
import pandas as pd
np.random.seed(50)

def random_walk(n_steps):
    step= 0.5 - np.random.random(n_steps)
    return np.cumsum(step)

date =pd.date_range('1/1/2019','12/31/2019',freq='D')
data = {'a':random_walk(365), 'b':random_walk(365)}
df=pd.DataFrame(data,index=date)
```
